```python
df_control = df[(df['variation'] == 'Control') & (df['process_step'] == 'confirm')]

df_control
```

```python
df_test = df[(df['variation'] == 'Test') & (df['process_step'] == 'confirm')]

df_test
```

```python
df_test = df_test.drop_duplicates(subset='client_id', keep='first')
df_test = df_test.drop_duplicates(subset='client_id', keep='first')
print("Unique IDs in df_test:", df_test['client_id'].nunique(), "Total rows:", len(df_test))
print("Unique IDs in df_control:", df_control['client_id'].nunique(), "Total rows:", len(df_control))
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

#H0 new design 'confirm' confirmation = old_confirmation
#H1 new design 'confirm' =! old design 'confirmation'

# Create a contingency table
data = {
    'Confirm': [len(df_test), len(df_control)],
    'Total': [len(df[df['variation'] == 'Test']), len(df[df['variation'] == 'Control'])]
}

contingency_table = pd.DataFrame(data, index=['Test', 'Control'])

# The total confirmations should be corrected if df_test and df_control only contain confirmations.
contingency_table['Non-Confirm'] = contingency_table['Total'] - contingency_table['Confirm']

# Perform Chi-squared test
chi2, p_value, dof, expected = chi2_contingency(contingency_table[['Confirm', 'Non-Confirm']])

# Print the results
print(f"Chi-squared Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Decision based on the p-value
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject the null hypothesis - There is a significant difference in the number of confirmations between the Test and Control groups.")
else:
    print("Fail to reject the null hypothesis - There is no significant difference in the number of confirmations between the Test and Control groups.")
```

```python
# checking compl rate incraese 5%

# Create a contingency table
data = {
    'Confirm': [len(df_test), len(df_control)],
    'Total': [len(df[df['variation'] == 'Test']), len(df[df['variation'] == 'Control'])]
}

# Convert total entries to total potential confirmations
contingency_table = pd.DataFrame(data, index=['Test', 'Control'])
contingency_table['Non-Confirm'] = contingency_table['Total'] - contingency_table['Confirm']

# Calculate completion rates
contingency_table['Completion Rate'] = contingency_table['Confirm'] / contingency_table['Total']

# Calculate the increase in completion rate
increase = (contingency_table.loc['Test', 'Completion Rate'] - contingency_table.loc['Control', 'Completion Rate']) / contingency_table.loc['Control', 'Completion Rate']
increase_percentage = increase * 100  # Convert to percentage

# Print the results
print("Contingency Table:")
print(contingency_table)
print(f"Increase in completion rate: {increase_percentage:.2f}%")

# Check if the increase meets or exceeds 5%
threshold = 5.0
if increase_percentage >= threshold:
    print("The observed increase in completion rate meets or exceeds the 5% threshold.")
else:
    print("The observed increase in completion rate does not meet the 5% threshold.")

# Optionally, perform Chi-squared test to validate statistical significance
chi2, p_value, dof, expected = chi2_contingency(contingency_table[['Confirm', 'Non-Confirm']])
print(f"Chi-squared Statistic: {chi2}")
print(f"P-value: {p_value}")
```

```python
#age hypothesis test
#H0 old people use new design


# Filter the DataFrame to include only users aged 50-80
age_filtered_df = df[(df['age'] >= 50) & (df['age'] <= 80)]

# Create a contingency table of counts for 'Test' and 'Control' variations
contingency_table = pd.crosstab(age_filtered_df['variation'], columns='count')

# Perform the chi-squared test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-squared Statistic: {chi2}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Decision based on the p-value
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis - There is a significant association between age group 50-80 and variation preference.")
else:
    print("Fail to reject the null hypothesis - There is no significant association between age group 50-80 and variation preference.")
```
